> **Research code accompanying the paper.** MolDeTr extracts δ, proton count, and couplings from real ¹H NMR spectra — including congested, strongly-coupled cases — and is largely field-agnostic (it works in Hz; tested on 80–600 MHz). Predictions can deviate outside its trained regime (unusual distortions, non-standard pulse sequences/processing, mixtures, or windows wider than 1200 Hz). `max J` is the dominant coupling per multiplet. See [docs/SCOPE.md](https://github.com/smidooo/MolDeTr/blob/main/docs/SCOPE.md).
